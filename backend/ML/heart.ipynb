```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import pickle
```

```python
# Load the user input data from the website (for example, using a form submission)
user_input = np.array([[63, 1, 3, 145, 233, 1, 0, 150, 0, 2.3, 0, 0, 1]])
user_data = np.array(user_input).reshape(1, -1)

# Preprocess user input data using the same StandardScaler used on training data
scaler = preprocessing.StandardScaler()
user_data_scaled = scaler.fit_transform(user_data.astype(float))

# Load the trained Naive Bayes classifier from the pickle file
with open("/kaggle/input/haertttttttt/naive_bayes_model.pkl", "rb") as file:
    nb = pickle.load(file)

# Predict user input data
user_prediction = nb.predict(user_data_scaled)

print("Predicted class for user input data is:", user_prediction[0])
```
